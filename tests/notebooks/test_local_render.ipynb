# Integration test notebook

This notebook is executed by `nbmake` (pytest) to verify that
`notebook2pdf` can convert a local notebook to PDF end-to-end.
It is marked `integration` and only runs when Quarto is on PATH.

```python
import pathlib
import shutil

if shutil.which('quarto') is None:
    raise RuntimeError('Quarto not found on PATH — skipping integration test')
```

```python
# When executed via nbconvert the kernel cwd is set to the directory passed
# as cwd in the subprocess call, which is the tmp_path containing this notebook.
THIS_NOTEBOOK = pathlib.Path.cwd() / 'test_local_render.ipynb'

from notebook_pdf import notebook2pdf
result = notebook2pdf(path=THIS_NOTEBOOK)
```

```python
assert result is not None, 'notebook2pdf returned None'
pdf = pathlib.Path(result)
assert pdf.exists(), f'PDF not found at {pdf}'
assert pdf.stat().st_size > 0, 'PDF is empty'
print(f'PDF produced: {pdf} ({pdf.stat().st_size} bytes)')
```
